```python
# sysmon.py
# ðŸš€ SysMon â€” Real-Time System Health Monitor

import psutil
import time
import platform
from datetime import datetime

def get_system_info():
    info = {
        "OS": platform.system(),
        "Version": platform.version(),
        "Processor": platform.processor(),
        "CPU Cores": psutil.cpu_count(logical=True),
        "RAM (GB)": round(psutil.virtual_memory().total / (1024 ** 3), 2),
    }
    return info

def get_metrics():
    return {
        "CPU Usage (%)": psutil.cpu_percent(interval=1),
        "Memory Usage (%)": psutil.virtual_memory().percent,
        "Disk Usage (%)": psutil.disk_usage("/").percent,
        "Network Sent (MB)": round(psutil.net_io_counters().bytes_sent / (1024 ** 2), 2),
        "Network Received (MB)": round(psutil.net_io_counters().bytes_recv / (1024 ** 2), 2),
    }

def display_metrics():
    print(f"\nðŸ”§ System Health @ {datetime.now().strftime('%H:%M:%S')}")
    for key, val in get_metrics().items():
        print(f"{key}: {val}")
    print("-" * 40)

if __name__ == "__main__":
    print("ðŸš€ SysMon â€” Real-Time System Health Monitor")
    print(get_system_info())
    print("\nMonitoring started... Press Ctrl+C to stop.\n")
    try:
        while True:
            display_metrics()
            time.sleep(2)
    except KeyboardInterrupt:
        print("\nðŸ›‘ Monitoring stopped.")
```
